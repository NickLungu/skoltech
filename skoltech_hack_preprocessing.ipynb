```python
def name_job_parse(name):
    retVal = ""
    if ("курьер" in name.lower() or "достав" in name.lower()):
        retVal += "курьер"
    elif ("экспедитор" in name.lower()):
        retVal += "экспедитор" 
    elif ("кладов" in name.lower()):
        retVal += "кладовщик"          

    if ("груз" in name.lower()):
        if len(retVal) != 0:
            retVal += " "      
        retVal += "грузовой"  

    if ("легк" in name.lower()):
        if len(retVal) != 0:
            retVal += " "      
        retVal += "легковой"      

    if ("на личном" in name.lower() or "на своем" in name.lower()):
        if len(retVal) != 0:
            retVal += " "
        retVal += "свое авто"                

    if ("авто комп" in name.lower() or "на наше" in name.lower()):
        if len(retVal) != 0:
            retVal += " "
        retVal += "авто компании"  
    
    if len(retVal) == 0 and "водитель" in name.lower():
        retVal = "Водитель"
    if len(retVal) == 0:
        retVal = name
    return retVal
```

```python
def shedule_job_parse(text):
    retVal = ""
    time = ["5/2","2/2","6/1","3/1","4/2","3/2"]
    for t in time:   
        if t in text:
            retVal = t
    return retVal
```

```python
data_jobs.head()
```

```python
data_jobs["name_coded"] = data_jobs.Name.apply(lambda x: name_job_parse(x))
data_jobs["shedule"] = data_jobs.Description.apply(lambda x: shedule_job_parse(str(x).lower()))
```

```python
data_jobs.drop(columns=["Name","Description"],axis=1,inplace=True)
```

```python
data_jobs.to_csv("data_jobs_fin.csv",encoding="utf-8-sig",index=False)
```
